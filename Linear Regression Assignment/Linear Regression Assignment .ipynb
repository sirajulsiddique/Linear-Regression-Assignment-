```python
from sklearn.model_selection import train_test_split as tts
```

```python
xtrain , xtest , ytrain, ytest = tts(x,y,train_size=.60,random_state=1) #test_size=.40
```

```python
lr.fit(xtrain ,ytrain)  #train the ML model ; xtrain=x ; ytrain=y
```
